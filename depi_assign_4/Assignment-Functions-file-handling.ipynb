## Q1: **Raising Custom Exceptions:**

You can define custom errors using the raise keyword.

Write a function that checks a withdrawal amount and raises an error if it is too high. Use try and except to handle the error. If the amount is more than 1000, raise a ValueError. Then call the function inside a try block and print an error message if the exception occurs.

- input
  check input int, str, float, ... etc
- amount
- account: 1000
- amount >1000 >>> error
- amount < 1000 >>> ok, remain


```python
#1st let's try and do this try and except task in a good way
#first we need to make the user enter the value the amount they want to withdraw
class OutOfRange(Exception):
      pass
def check_withdrawal_amount(user_input):
        try:
            amount = int(user_input)
            if amount > 1000:
                    raise OutOfRange
            else:
                    return amount
        except ValueError:
                print("please enter a valid input.")
                return 'a' #im doing this to make something returned that can be used in order to control the loop at the while True to simulate a real atm so the user can keep trying
                
        except OutOfRange: 
                print("The amount asked for is too large!!")
                return None
    
while True:
    amount = input("enter the amount you want to withdraw")
    if amount.upper() == "EXIT":
        print("You are exiting the machine")
        break
    amount_valid = check_withdrawal_amount(amount)
    if amount_valid == None:
        print("please enter a valid input")
        continue
    else:
          print(f"you have drawn {amount_valid} from your bank account succefully..")



```

## Q2: **simple pipeline with three functions**

- read_file → reads text/ csv
- process_text → does something simple
- write_file → writes processed output into a processed/ directory

How it works

- Put an input file: data.txt
  Run:
- run_pipeline("data.txt") or run_pipeline("data.csv")

Output appears in:
processed/data.txt


```python
def extract(data_source):
    with open(data_source, 'r+') as f:
        data = f.read()
        return data

def transform(data):
    transformed_data = list(data.split())
    transformed_list = []
    for word in transformed_data:
        transformed_list.append(word.upper())
    return transformed_list

def load(transformed_data):
    print(transformed_data)
    return transformed_data
```
