#### Q_5: Fill missing values in the Age column with the mean age then Group the data by Pclass and calculate the average age for each class

```python
#now lets calculate the average afe for each class by Pclass 
df['Pclass'].unique() #this is an example to see the array of classes
df['Pclass'].value_counts() #this gets the count of each class
df.groupby('Pclass', as_index=False)['Age'].mean() #as index false make it appear as it is not an index
```

```python
mean_age = df['Age'].mean()
print(mean_age)
df['Age'] = df['Age'].fillna(mean_age)
df['Age'] = (df['Age'] * 2).round() / 2
#i searched and did this to make the format look good and look like the age before i fill it with the mean actaully
pd.options.display.float_format = '{:g}'.format
df.head(40)
```
