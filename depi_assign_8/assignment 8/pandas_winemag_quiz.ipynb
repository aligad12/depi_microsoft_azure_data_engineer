## ðŸ”¹ Section 2: Filtering & Selection


```python
df.head(10)
```

```python
# Q4: Wines from Italy with points >= 90
df.head(10)
#so we need to filter countries and select italy
filtered_data_frame = df[(df['country'] == 'Italy') & (df['points'] >=90)]
filtered_data_frame.head(10)
# as we can see we were able to filter it, i struggle while writing this but learnt that i need to put each coniditon between brackets to avoid errors
```

```python
# Q5: Cheapest wine with points >= 95
wines_high_points = df[df['points'] >= 95]
minimum_price = wines_high_points['price'].min()
print(minimum_price)
cheapest_wine = wines_high_points[wines_high_points['price'] == minimum_price]
pd.set_option('display.max_columns',None)
print(cheapest_wine.to_string(index=False)) 
#there is 5 with price 20 thats why the data is appearing like this
```

```python
# Q6: Top 5 countries by number of wines
df['country'].unique()
df['country'].value_counts().to_frame()
#now we want to make a data frame with the max number of count, which contains 5 countries
top_5names = ['US','France','Italy','Spain','Portugal'] # these are top 5 countries with number of wines, and then i filtered also the data frame to contain them only
top = df[df['country'].isin(top_5names)]
top.head(5)
```

## ðŸ”¹ Section 3: Grouping & Aggregation


```python
# Q7: Average price per country
df.groupby('country')['price'].mean()
#this is how to calculate mean price per country
```

```python
# Q8: Top 5 varieties by average points (min 100 reviews)
groups_with_reviews = df.groupby('variety').filter(lambda x: len(x) >= 100)
calculated_means = groups_with_reviews.groupby('variety')['points'].mean()
sorted_values = calculated_means.sort_values(ascending=False)
sorted_values.to_frame().head(5) # i did this to print it in a data frame shape
```

```python
# Q9: Country stats
df.head(3)
countries = df.groupby('country')
countries.describe()
```

## ðŸ”¹ Section 4: Feature Engineering


```python
# Q10: Create price_per_point column
df.head(10)
df['price_per_point'] = df['price']/df['points']
df.head(3)
```

```python
# Q11: Best value wines (points >= 90)
best_value_wines = df[df['points'] >=90 ]
best_value_wines.head(5)
```

```python
# Q12: Fill missing taster names
best_value_wines = best_value_wines.copy() # when i did not do this i got a warning so i searched how to fix this warning, and it told me when you make a data frame( filtered one) from another data frame
#this causes pandas to get confused so the best is to tell pandas that you made a copy of the filtered one and started modifying on it, and its better to use .loc() here also after doing this
best_value_wines.loc[:,'taster_name'] = best_value_wines['taster_name'].fillna("missing")
best_value_wines.head(5)

```

## ðŸ”¹ Section 5: Text Operations


```python
# Q13: Descriptions containing 'vanilla' or 'oak'
#to do this element wise operation
filtered_data_frame =df[df['description'].str.contains('vanilla|oak',case=False,na=False)] #we used a regex here
filtered_data_frame.head(5)
```

```python
# Q14: Top 5 tasters by number of reviews
counts = df.groupby('taster_name').size()
top_5 = counts.sort_values(ascending=False).head(5)
top_5.head().to_frame()
```

## ðŸ“Š Section 6: Simple Visualization


```python
df = pd.read_csv('winemag-data-130k-v2.csv')
```

```python
# Q15: Bar chart - top 10 countries
#lets create a bar chart for top 10 countries
#top 10 countries by points
import matplotlib.pyplot as plt
top_10 = df.groupby('country')['points']
top_10_final = top_10.mean().sort_values(ascending=False).head(10)
top_10_final.plot(x = 'country',y='points',kind='barh',color='orange')
plt.xlabel('Mean Points')
plt.ylabel('Country')
plt.title('Top 10 countries by average Points')
plt.show()
```

```python
# Q16: Histogram of prices below 200
prices_below_200 = df[df['price'] < 200]
prices_below_200 = prices_below_200.copy()
prices = prices_below_200['price']
prices.plot(kind='hist',bins= 20) #this makes it more detailed view when i increase the number of bins
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Price Below 200')
plt.show()
```
