3.1 Guest List: If you
could invite anyone, living or deceased,
to dinner, who would you invite? Make a
list that includes at least three people
you’d like to invite to dinner. Then use
your list to print a message to each
person, inviting them to dinner.


```python
living_persons= ["Ali","Mohamed","Mahmoud","Ibrahim"]
deceased_persons = ["Hossam","Hassan","barakat","islam"]
```

```python
invited_persons = []
invited_persons.append(living_persons[0])
invited_persons.append(living_persons[1])
invited_persons.append(deceased_persons[0])
for person in invited_persons:
    print(f"Hello {person} you are invited to the dinner :)")
```
