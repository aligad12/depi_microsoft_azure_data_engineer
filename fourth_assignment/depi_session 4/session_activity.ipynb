```python
#first method to get even number:
def is_even(number):
    if number % 2 == 0:
        return True
    else:
        return False
```

```python
def get_first_character(word):
    if len(word) > 0:
        return word[0]
    else:
        return None # for empty string 

print(get_first_character("Hello"))
print("")
```

```python
words = ["apple","banana","cherry","date","elderberry"]

def get_longest_string(words):
    max = len(words[0])
    if len(words) !=0:
        for word in words:
            if len(word) > max:
                max = len(word)
                selected_word = word
        return selected_word
    else:
        return None

print(get_longest_string(words))
```
